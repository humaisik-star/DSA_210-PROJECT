## Relationship Analysis

- There appears to be a positive relationship between hours studied and performance.
- Previous scores seem strongly correlated with performance index.
- Sleep hours may have a moderate effect on performance.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv("../data/raw/Student_Performance.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Sleep Hours", y="Performance Index")
plt.title("Sleep Hours vs Performance Index")
plt.xlabel("Sleep Hours")
plt.ylabel("Performance Index")
plt.savefig('../figures/sleep_vs_performance.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Hours Studied", y="Performance Index")
plt.title("Hours Studied vs Performance Index")
plt.xlabel("Hours Studied")
plt.ylabel("Performance Index")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Previous Scores", y="Performance Index")
plt.title("Previous Scores vs Performance Index")
plt.xlabel("Previous Scores")
plt.ylabel("Performance Index")
plt.show()
```

## Correlation Analysis

- Previous Scores and Hours Studied show strong correlation with Performance Index.
- Sleep Hours has a weaker but noticeable relationship.

```python
sns.boxplot(x="Extracurricular Activities", y="Performance Index", data=df)
plt.title("Extracurricular Activities vs Performance")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Extracurricular Activities

- Students participating in extracurricular activities may show differences in performance.
- Extracurricular Activities indicates some relationship with Performance Index.

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Extracurricular Activities", y="Performance Index")

plt.title("Performance Index by Extracurricular Activities")
plt.xlabel("Extracurricular Activities")
plt.ylabel("Performance Index")

plt.show()
```

```python
df.groupby("Extracurricular Activities")["Performance Index"].mean()
```

```python
df.groupby("Extracurricular Activities")["Performance Index"].describe()
```

## Extracurricular Activities Analysis

Students who participate in extracurricular activities may show differences in academic performances. 

The boxplot suggests that there is a difference in the distribution of Performance Index between students who participate and those who do not.

The group comparison shows that the average performance differs between the two groups, indicating that extracurricular involvement may have an impact on academic outcomes.

```python
sns.violinplot(data=df, x="Extracurricular Activities", y="Performance Index")
plt.title("Detailed Distribution by Extracurricular Activities")
plt.show()
```
