## Initial Observations

- The dataset contains 10,000 entries and 6 columns.
- There are no missing values in the dataset.
- The target variable appears to be "Performance Index".
- Features include study habits, sleep duration, and previous performance.

```python
# Distribution of Performance Index
sns.histplot(df["Performance Index"], kde=True)
plt.title("Distribution of Performance Index")
plt.show()
```

```python
# Distribution of Sleep Hours
sns.histplot(df["Sleep Hours"], kde=True)
plt.title("Distribution of Sleep Hours")
plt.show()
```

## Distribution Analysis

- Performance Index appears to be normally distributed.
- Sleep Hours show a range of values indicating variation among students.

```python
sns.scatterplot(x="Sleep Hours", y="Performance Index", data=df)
plt.title("Sleep Hours vs Performance Index")
plt.show()
```

```python
sns.scatterplot(x="Hours Studied", y="Performance Index", data=df)
plt.title("Hours Studied vs Performance Index")
plt.show()
```

```python
sns.scatterplot(x="Previous Scores", y="Performance Index", data=df)
plt.title("Previous Scores vs Performance Index")
plt.show()
```

## Relationship Analysis

- There appears to be a positive relationship between hours studied and performance.
- Previous scores seem strongly correlated with performance index.
- Sleep hours may have a moderate effect on performance.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
